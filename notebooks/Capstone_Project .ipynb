```python
#identifying null values
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce') #converting into numeric datatype
df.fillna(np.nan, inplace=True) #replacing non-numeric values with Nan
df['TotalCharges'] = df['TotalCharges'].astype(np.float64) #changing the datatype to float64
print(df.isnull().sum())
```

```python
# Drop rows with NaN in 'TotalCharges'
df.dropna(subset=['TotalCharges'])
df[df['TotalCharges'].isnull()]
```

```python
filtered_index = df[df['tenure'] == 0].index
# Use filtered_index for further analysis
df.drop(filtered_index, axis=0, inplace=True) #droping 11 missing values in tenure
filtered_index
```
